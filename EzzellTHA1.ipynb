**Introduction to the data sources:**
The following data sets display asylum data from 2018.


*  ** "1.Field Office Overview.xlsx" **- This data set is an excel file which shows Custom and Border Patrol organizational data including field offices, the sectors which correspond to each field office, the name of the ports of entry for each field office, the Border Patrol designated port code for each port of entry and their geographic location. The zipped file can be found on canvas and was created using information from the Customs and Border Patrol webite.
*   **"2. Indmissables2018.xlsx**- This data set is an excel file showing the number of asylum seekers who crossed at an official port of entry on the US southern border in 2018. The data is organized by the Border Patrol sector where the asylum seekers crossed on the Southern border. "Family Unit" refers to the number of individuals in a family who cross. "UAC" refers to the number of unaccompanied minors who crossed at the port of entry. The original data can be found at the following link: https://www.cbp.gov/newsroom/stats/ofo-sw-border-inadmissibles. The excel file for the data is uploaded as a zipped file one Canvas.


*   **"3.Apprehensions 2018.xlsx"**-This data set is an excel file showing the number of asylum seekers who crossed in between ports of entry on the US southern border in 2018. The data is organized by the Border Patrol sector where the asylum seekers crossed on the Southern border. "Family Unit" refers to the number of individuals in a family who cross. "UAC" refers to the number of unaccompanied minors who crossed at the port of entry. The original data can be found at the following link: https://www.cbp.gov/newsroom/stats/usbp-sw-border-apprehensions  The excel file for the data is uploaded as a zipped file one Canvas.





```
import pandas as pd
inadmin2018=pd.read_excel("2. Inadmissables 2018.xlsx")
apprehen2018=pd.read_excel("3. Apprehensions 2018.xlsx")
FieldOffice=pd.read_excel("1. Field Office Overview.xlsx")


print(inadmin2018)
print(apprehen2018)
print(FieldOffice)
```

```
apps_in=inadmin2018.merge(apprehen2018,on="Field Office")
print(apps_in)
asylum2018=apps_in.merge(FieldOffice,on="Field Office")
print(asylum2018)
```

```
asylum2018.to_excel("asylum2018.xlsx")
```
